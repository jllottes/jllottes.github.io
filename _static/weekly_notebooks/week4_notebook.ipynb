# The Collatz Conjecture

Consider the function
$$
f(n) = 
\begin{cases}
    3n+1 & \text{if $n$ is odd}
    \\
    n/2 & \text{if $n$ is even}
\end{cases}
$$

For any starting $n$, we generate a sequence $\{a_i\}$ where
$$
a_0 = n, \qquad a_{i+1} = f(a_i) \quad \text{for i = 0,1,2,}\cdots
$$

```python
def f(n):
    if n%2 == 1:
        return 3*n+1
    else:
        return n//2
```

The Collatz conjecture states that for any starting positive integer $n$, the sequence eventually reaches 1.

**Exercise.** Define a function `collatz(n)` that generates the sequence of values $\{a_i\}$ where $a_0 = n$ and $a_{i+1} = f(a_i)$, stopping when the sequence reaches 1.

```python
def collatz(n):
    orbit = [n]
    while True:
        ai = orbit[-1] # most recent number in the sequence
        fai = f(ai)    # find the next entry in the sequence
        orbit.append(fai)
        if fai == 1:
            break
    return orbit
```

We can clean this up slightly:

```python
def collatz(n):
    orbit = [n]
    while not orbit[-1] == 1:
        orbit.append(f(orbit[-1]))
    return orbit
```

```python
collatz(32)
```

Let's look at these orbits for many numbers:

```python
orbits = []

N = 10**5

nlist = [n for n in range(1,N)]

for n in nlist:
    orbits.append(collatz(n))
```

A tuple of integers $(a,b,c)$ is a **Pythagorean triple** if $a^2 + b^2 = c^2$.

Let's start with a function that takes in an integer and checks if it is a perfect square.

```python
from math import sqrt

def is_square(n):
    if sqrt(n) == round(sqrt(n)):
        return True
    else:
        return False
```

```python
is_square(4)
```

This looks like it's working, but it's always risky to check whether two floats are equal to one another. Another approach that is slightly safer:

First, compute $\sqrt{n}$. Then take the integer part of $\sqrt{n}$. Then square the result and check if it is equal to $n$.

```python
def is_square(n):
    if int(sqrt(n)) ** 2 == n:
        return True
    else: return False
```

```python
is_square(4)
```

There is still a possible complication if Python evaluated something like `sqrt(9.0) = 2.99999999`. We can get around this by adding say `0.5` to `sqrt(n)` before truncating to make up for any slight under estimation of the square root.

```python
def is_square(n):
    if int(sqrt(n) + 0.5) ** 2 == n:
        return True
    else: return False
```

```python
is_square(9)
```

## The greatest common divisor

The greater common divisor of two integers $a$ and $b$ is the greatest integer that divides both $a$ and $b$.
We denote it by $\gcd(a,b)$.

If $\gcd(a,b)=1$, then we say that $a$ and $b$ are relatively prime.

One method of finding $\gcd(a,b)$ is to brute-force check every possible divisor, then take the maximum of all divisors.a

```python
def gcd(a,b):
    divisors = []
    for d in range(1,min(a,b)+1):
        if a % d == 0 and b % d == 0:
            divisors.append(d)
    return max(divisors)
```

```python
gcd(12,18)
```

## The Euclidean algorithm for computing the gcd

(see [wikipedia article](https://en.wikipedia.org/wiki/Euclidean_algorithm#:~:text=In%20mathematics%2C%20the%20Euclidean%20algorithm,them%20both%20without%20a%20remainder.&text=When%20that%20occurs%2C%20they%20are,of%20the%20original%20two%20numbers.))

* Start with two integers $a$, $b$
* Subtract the smaller number from the larger and replace the larger with this new number
* Repeat the above step until one of the numbers is 0
* the other number is the $\gcd(a,b)$.
