```python
import matplotlib.pyplot as plt
```

Recall: We can generate a list of $N+1$ evenly spaced points between $a=-\pi$ and $b=\pi$ as follows:

```python
from math import pi
```

```python
N = 1000

x = [-pi + 2*pi*i/N for i in range(N+1)]
```

Let's plot the $\sin(x)$ and the $\cos(x)$.

```python
from math import sin,cos
```

```python
sinx = [ sin(xx) for xx in x ]
cosx = [ cos(xx) for xx in x ]

plt.plot(x,sinx,'r-',label='$\sin x$')
plt.plot(x,cosx,'b-',label='$\cos x$')

plt.legend()
```

Here, the $\sin$ and the $\cos$ curves are on the same set of axes. What if we want to plot them on separate axes?

We can manually construct a figure in `matplotlib` using `plt.figure()`.

```python
plt.figure()
plt.plot(x,sinx,'r-',label='$\sin x$')
plt.ylabel('$y$')

plt.figure()
plt.plot(x,cosx,'b-',label='$\cos x$')
plt.legend()

plt.xlabel('$x$')
```

Notice: the legend only appears on the second figure. When we call `plt.figure`, a new figure becomes *active*. All future calls to `plt` will apply to the new figure.

An alternative approach is to create one figure with several axes.

## Subplots

The matplotlib `subplot()` function creates a matplotlib figure consisting of several subplots. The syntax is:

`plt.subplot(r,c,n)` where:

* `r` = the number of rows in the grid
* `c` = the number of columns in the grid
* `n` = the number of subplot in the grid

Let's plot the $\sin$ and $\cos$ curves as subplots that are side-by-side.

**Note:** matplotlib does not use zero-based indexing for the subplot number.

```python
plt.figure()

plt.subplot(1,2,1)
plt.plot(x,sinx,'r-',label='$\sin x$')
plt.title('$\sin x$')

plt.subplot(1,2,2)
plt.plot(x,cosx,'b-',label='$\cos x$')
plt.title('$\cos x$')
```

We can modify the shape of the figure by adding the keyword `figsize = (w,h)` to the `plt.figure()` command:

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(x,sinx,'r-',label='$\sin x$')
plt.title('$\sin x$')

plt.subplot(1,2,2)
plt.plot(x,cosx,'b-',label='$\cos x$')
plt.title('$\cos x$')
```

We can change our grid shape to plot these vertically instead.

```python
plt.figure(figsize=(10,5))

plt.subplot(2,1,1)
plt.plot(x,sinx,'r-',label='$\sin x$')
plt.title('$\sin x$')

plt.subplot(2,1,2)
plt.plot(x,cosx,'b-',label='$\cos x$')
plt.title('$\cos x$')
```

We can see that the title of the lower axis is colliding with the upper axis.
We can adjust the spacing using `plt.subplots_adjust`:

```python
plt.figure(figsize=(10,5))

plt.subplot(2,1,1)
plt.plot(x,sinx,'r-',label='$\sin x$')
plt.title('$\sin x$')

plt.subplot(2,1,2)
plt.plot(x,cosx,'b-',label='$\cos x$')
plt.title('$\cos x$')

plt.subplots_adjust(hspace = .4)
# wspace = ... can be used to specify the horizontal padding between subplots
# hspace = ... can be used to specify the vertical padding between subplots
```

We can build larger arrays of subplots. Let's build a $2\times 2$ grid of subplots.

```python
plt.figure(figsize = (8,5))

plt.subplot(2,2,1) # create the first subplot in a 2x2 grid
plt.plot(x,sinx)
plt.title('(2,2,1)')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.subplot(2,2,2) # create the second plot in a 2x2 grid
plt.plot(x,cosx)
plt.title('(2,2,2)')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.subplot(2,2,3) # create the third plot in a 2x2 grid
x2 = [ xx**2 for xx in x ]
plt.plot(x,x2)
plt.title('(2,2,3)')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.subplot(2,2,4) # create the fourth plot in a 2x2 grid
x3 = [ xx**3 for xx in x ]
plt.plot(x,x3)
plt.title('(2,2,4)')
plt.xlabel('$x$')
plt.ylabel('$y$')
```

We could use `plt.subplots_adjust` to try to fix the spacing between plots. Alternatively, we can use `plt.tight_layout()` to have matplotlib try to fix the spacing on its own.

```python
plt.figure(figsize = (8,5))

plt.subplot(2,2,1) # create the first subplot in a 2x2 grid
plt.plot(x,sinx)
plt.title('(2,2,1)')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.subplot(2,2,2) # create the second plot in a 2x2 grid
plt.plot(x,cosx)
plt.title('(2,2,2)')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.subplot(2,2,3) # create the third plot in a 2x2 grid
x2 = [ xx**2 for xx in x ]
plt.plot(x,x2)
plt.title('(2,2,3)')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.subplot(2,2,4) # create the fourth plot in a 2x2 grid
x3 = [ xx**3 for xx in x ]
plt.plot(x,x3)
plt.title('(2,2,4)')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.tight_layout() # have matplotlib auto set the spacing
```

### Back to subplots:

We can mix and match grid layouts as long as our plots don't overlap.
As an example, we can create a column on the left with 2 rows, and a subplot in the right column spanning both rows.

```python
plt.figure()

plt.subplot(2,2,1)
plt.plot(x, sinx)
plt.title('(2,2,1)')

plt.subplot(2,2,3)
plt.plot(x, cosx)
plt.title('(2,2,3)')

plt.subplot(1,2,2)
plt.plot(x, x2)
plt.title('(1,2,2)')

plt.tight_layout()
```

When we call `plt.subplot()` as above, we generate generate a new subplot that becomes *active*. Future plot commands apply to that subplot until a new subplot becomes active.

There may be times where we want to go back to an earlier subplot and make changes. To do so, we can store our subplots as variables and then recall them with the `plt.axes()` command.

```python
plt.figure()

ax1 = plt.subplot(2,2,1)  # store the first subplot as ax1
plt.plot(x,sinx)
plt.title('(2,2,1)')

ax2 = plt.subplot(2,2,2) # store the second subplot as ax2
plt.plot(x,cosx)
plt.title('(2,2,2)')

ax3 = plt.subplot(2,1,2) # store the third subplot as ax3
plt.plot(x,x2)
plt.title('(2,1,2)')

plt.xlabel('$x$')  # add an x-label to ax3, since that is the active subplot

plt.axes(ax1) # re-activate ax1
plt.xlabel('$x$')

plt.axes(ax2) # re-active ax2
plt.ylabel('$y$')

plt.tight_layout()
```

The `plt.axes` command can also be used to generate free-form subplots.
Synatx: `plt.axes([x,y,w,h])` where:
* `(x,y)` = coordinates of the lower left corner of the axes object
* `w` = width of the axes
* `h` = height of the axes

```python
plt.figure(figsize=(8,5))

ax1 = plt.axes([.2,.1,.7,.5])
ax2 = plt.axes([.5,.4,.5,.5])

plt.axes(ax1)
plt.plot(x,sinx,'b-')

plt.axes(ax2)
plt.plot(x,cosx,'r-')
```

## NumPy

NumPy stands for numerical Python, and is one of the most widely used packages in Python.

```python
import numpy as np
```

### numpy arrays:

A key feature of numpy are arrays, which resembles lists.

A list can be converted to a numpy array using the `np.array()` command.

```python
mylist = [1,2,3]
print(mylist)
```

```python
myarray = np.array(mylist)
print(myarray)
```

Let's check the type of these objects using `type()` function:

```python
type(mylist)
```

```python
type(myarray)
```

Let's compare some mathematical operations performed on lists versus arrays

```python
print(mylist + mylist)
```

```python
print(myarray + myarray)
```

Adding lists concatenates them.

Adding arrays (of appropriate shapes) perform element-wise addition.

We can also add numbers to arrays, which is done element-wise. The cannot be done for lists:

```python
print(mylist + 1)
```

```python
print(myarray)
print(myarray + 1)
```

We have similar behavior with multiplication by an integer:

```python
print(mylist*4)
print(myarray*4)
```

We can't multiply two lists together, but we can multiply two (appropriately shaped) arrays together, which is performed element-wise:

```python
print(mylist*mylist)
```

```python
print(myarray*myarray)
```

We can exponentiate arrays also:

```python
print(myarray**2)
```

**Note:** Two NumPy arrays must have compatible shape to perform operations involving them:

```python
myarray
print(myarray)
```

```python
myarray2 = np.array([4,5,6,7])
print(myarray2)
```

```python
myarray + myarray2
```

We can use slicing to access elements from arrays just as we do with lists:

```python
myarray[0]
```

```python
myarray[-1]
```

```python
myarray2[1:3]
```

We can generate an array of equally spaced points using the `np.linspace` command. The syntax is `np.linspace(a,b,N)` where:
* `a` is the first $x$-value
* `b` is the last $x$-value
* `N` is the number of grid points

```python
N = 1000
x = np.linspace(-pi,pi,N+1)
```

```python
cos(x)
```

I can't use the `cos` function from the math module on numpy arrays. Instead, we should use the `np.cos` function, i.e. the cosine function that is supplied by numpy.

```python
y = np.cos(x)
```

Matplotlib is friendly to numpy arrays:

```python
plt.plot(x,y)
```

**Note:** Operations in numpy can be much faster than in pure Python:

```python
from time import time
```

```python
N = 10**7
x = np.linspace(-pi,pi,N+1)

t0 = time()
[sin(xx) for xx in x]
t1 = time()
print(t1 - t0)

t0 = time()
np.sin(x)
t1 = time()
print(t1 - t0)
```
